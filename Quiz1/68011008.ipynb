## PART NUMPY

```python
LIST = [10, 20, 30, 40]
LIST
```

### Q1
1. แปลง list เป็น ndarray โดยการใช้คำสั่ง `np.array(LIST)`

```python
ans_1 = np.array(LIST)
ans_1
```

### Q2
1. shape จากข้อ 1 ยังไม่ตรง (1, 4) เลยต้องใช้ `.reshape((shape))` เพื่อให้ shape ตรง (2, 2)
2. จากนั้นทำการ - แบบ element wise
3. จากนั้นทำการ สร้าง random ndarray ด้วย `np.random.random((shape))`

```python
np.random.seed(42)
```

```python
ans_2 = ans_1.reshape((2, 2)) - 5 * np.array([[1, 1], [1, 1]]) - np.random.random((2, 2))
ans_2
```

### Q3
1. ทำการ slicing ด้วย `[:, 0]` (: หมายถึงทั้งหมด)
2. ทำการต่อด้วยการ `np.concat([arr1, arr2, ... arr3], dim=0)` (dim=0 คือชั้นนอกสุด)

```python
ans_3 = ans_2[: , 0]
ans_3 = ans_3.reshape((2, 1))
ans_3 = np.concat([ans_3, ans_3], axis=0)
ans_3
```

## PART PYTORCH

### Q4
1. ทำการสร้าง tensor ด้วยคำสั่ง `torch.tensor(LIST)`

```python
ans_4 = torch.tensor(LIST)
ans_4
```

### Q5
1. ทำการแปลง shape ด้วยคำสั่ง `.reshape(shape)`
2. ทำการลบแบบ element wise กับ random tensor จากคำสั่ง `torch.rand((shape))`

```python
torch.manual_seed(42)
```

```python
ans_5 = ans_4.reshape(2, 2)
ans_5 = ans_5 - 2 * torch.tensor([[1, 0], [0, 1]]) + torch.rand((2, 2))
ans_5
```

### Q6
1. slice ด้วย `[1, :]` (: คือทั้งหมด)
2. ทำการต่อ tensor ด้วย `torch.cat((tensor1, tensor2, ..., tensorn), dim=0)`

```python
ans_6 = ans_5[1, :]
ans_6 = torch.cat((ans_6, ans_6), dim=0)
ans_6
```

### Q7
1. ทำการแปลง ans_3 เป็น tensor ด้วย `torch.tensor(ndarray, dtype=float)`
2. reshape ให้ทำการ matmul ได้
2. ทำการ mat mul ด้วย operator `@`

```python
ans_7 = torch.tensor(ans_3, dtype=torch.float64)
ans_6_2d = ans_6.reshape(1, 4)
ans_7 = ans_7 @ ans_6_2d.double()
ans_7
```
